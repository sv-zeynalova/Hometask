```python
# Сколько вакансий для аналатика данных?
result_1 = vacancy.vactitle.str.lower().str.contains('аналитик данных')
result_2 = vacancy.vactitle.str.lower().str.contains('data analyst')
DA=vacancy[result_1 | result_2]
DA.shape
```

```python
# Сколько вакансий для аналитика данных с использованием Python?

result_3 = vacancy.vactitle.str.lower().str.contains('аналитик данных')
result_4 = vacancy.vacdescription.str.lower().str.contains('python')
vacancy[result_3 & result_4].shape
```

```python
#Сколько вакансий, которые вам нравятся?
t1 = vacancy.vacdescription.notna()
t2 = vacancy.vactitle.str.lower().str.contains('аналитик данных')
t3= vacancy.vactitle.str.lower().str.contains('data analyst')
t4 = vacancy.vacdescription.str.lower().str.contains('обучение')
my_list=vacancy[t1&t2|t3&t4]
my_list.shape
```

```python
 #Насколько свежие эти вакансии? 12.03.2021-23.10.2020
my_list.reset_index().sort_values(by='vacdate', ascending=False)
```
